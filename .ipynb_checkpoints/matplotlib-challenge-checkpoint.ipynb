```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Instructions/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_data_to_load = "Instructions/Pymaceuticals/data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drug_data_complete = pd.merge(trial_data, drug_data, how = "left", on = ["Mouse ID"])

# Display the data table for preview
drug_data_complete.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_drug_data = drug_data_complete.groupby(["Drug", "Timepoint"])
grouped_means = grouped_drug_data.mean()
# Convert to DataFrame
mean_tumorv_df = grouped_means.drop(columns = ["Metastatic Sites"])
# Preview DataFrame
mean_tumorv_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_sem = grouped_drug_data.sem()
sem_tumorv_df = grouped_sem.drop(columns = ["Mouse ID", "Metastatic Sites"])
# Convert to DataFrame
sem_tumorv_df2 = sem_tumorv_df.pivot_table("Tumor Volume (mm3)", "Timepoint", "Drug")
# Preview DataFrame
sem_tumorv_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumorv_df2 = mean_tumorv_df.pivot_table("Tumor Volume (mm3)", "Timepoint", "Drug")
# Preview that Reformatting worked
mean_tumorv_df2.head()
```

```python
# Generate the Plot (with Error Bars)
time = mean_tumorv_df2.index
drugs = mean_tumorv_df2.columns

fig, ax = plt.subplots()

#To print all drugs in plot
# for i in drugs:
#     spec_drug_data = mean_tumorv_df2.loc[:,i]
#     error = sem_tumorv_df2.loc[:,i]    
#     plt.scatter(time, spec_drug_data, c=["b"])
#     ax.errorbar(time, spec_drug_data, yerr = error, 
#                 fmt="o", color="b", alpha=0.5, label = i)
#     plt.legend(loc="upper left", fontsize="small", fancybox=True)

#To print required drugs
cap = mean_tumorv_df2.loc[:,"Capomulin"]
cap_error=sem_tumorv_df2.loc[:,"Capomulin"]
plt.scatter(time, cap, c=["b"], marker = "o")
ax.errorbar(time, cap, yerr = cap_error, fmt="o", color="b", alpha=0.5, label = "Capomulin")

inf = mean_tumorv_df2.loc[:,"Infubinol"]
inf_error=sem_tumorv_df2.loc[:,"Infubinol"]
plt.scatter(time, inf, c=["r"], marker = "^")
ax.errorbar(time, inf, yerr = inf_error, fmt="^", color="r", alpha=0.5, label = "Infubinol")

ket = mean_tumorv_df2.loc[:,"Ketapril"]
ket_error=sem_tumorv_df2.loc[:,"Ketapril"]
plt.scatter(time, ket, c=["g"], marker = "x")
ax.errorbar(time, ket, yerr = ket_error, fmt="x", color="g", alpha=0.5, label = "Ketapril")

pla = mean_tumorv_df2.loc[:,"Placebo"]
pla_error=sem_tumorv_df2.loc[:,"Placebo"]
plt.scatter(time, pla, c=["black"], marker = "*" )
ax.errorbar(time, pla, yerr = pla_error, fmt="*", color="black", alpha=0.5, label = "Placebo")

plt.legend(loc="upper left", fontsize="small", fancybox=True)

plt.grid()
plt.title("Tumor Growth During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
#Save the Figure
plt.show()
plt.savefig('Tumor Growth During Treatment.png')
```

```python
# Calculate the percent changes for each drug
df = drug_data_complete.pivot_table("Tumor Volume (mm3)", "Timepoint", "Drug")
first_row = df.loc[0]
last_row = df.loc[45]
pct_change = (last_row-first_row)/first_row *100
# Display the data to confirm
pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_change_t = tuple(pct_change)
```

```python
# Splice the data between passing and failing drugs
pos_drugs = pct_change[pct_change >= 0]
neg_drugs = pct_change[pct_change <= 0]

# Orient widths. Add labels, tick marks, etc. 
x_axis1 = np.arange(len(pos_drugs))
x_axis2 = [-1,-2]

plt.xlim()
plt.ylim()

tick_locations = np.arange(-2,len(drugs)-2)
drugs_order = list(neg_drugs.index) + list(pos_drugs.index)

plt.xticks(tick_locations, drugs_order, rotation = 90)
plt.bar(x_axis1, pos_drugs)
plt.bar(x_axis2, neg_drugs)

# Give our chart some labels and a tile
plt.title("Tumor Growth Over 45 Days")
plt.xlabel("Drugs")
plt.ylabel("% Change of Tumor Volume")

# Use functions to label the percentages of changes

# Call functions to implement the function calls
#plt.text(x = tick_locations, y=0, s = drugs)
plt.grid()
plt.show

# Save the Figure
plt.savefig('Tumor Growth Over 45 Days.png')
# Show the Figure
fig.show()
```
